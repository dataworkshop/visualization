```python
import matplotlib.pyplot as plt
import numpy as np

from bokeh import mpl
from bokeh.plotting import output_notebook, show
```

```python
output_notebook(hide_banner=True)

def f(x):
    return np.pi*np.sin(x) + np.pi*np.cos(np.pi * x) + np.sin(x + np.pi)

def g(x):
    x /= 4
    return np.pi*np.cos(x) + np.pi*np.sin(np.pi * x) + np.cos(x + np.pi)


x = np.linspace(-3*np.pi, 3*np.pi, 500)
a = f(x)
b = g(x)

plt.title('Matplotlib Figure in Bokeh')
plt.plot(x, a, 'r-', marker='o')
plt.plot(x, b, 'g-x', linestyle='-.')
```

```python
show(mpl.to_bokeh())
```
